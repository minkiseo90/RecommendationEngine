## Text Nomalization (텍스트 정규화)

```python
replacement_patterns = [
    (r'won\'t', 'will not'),
    (r'can\'t', 'cannot'),
    (r'i\'m', 'i am'),
    (r'I\'m', 'i am'),
    (r'ain\'t', 'is not'),
    (r'(\w+)\'ll', '\g<1> will'),
    (r'(\w+)n\'t', '\g<1> not'),
    (r'(\w+)\'ve', '\g<1> have'),
    (r'(\w+)\'s', '\g<1> is'),
    (r'(\w+)\'re', '\g<1> are'),
    (r'(\w+)\'d', '\g<1> would'),
    (r'(\w+)\'s', '\g<1> is')
]

class RegexpReplacer(object):
    """ Replaces regular expression in a text.
    >>> replacer = RegexpReplacer()
    >>> replacer.replace("can't is a contraction")
    'cannot is a contraction'
    >>> replacer.replace("I should've done that thing I didn't do")
    'I should have done that thing I did not do'
    """
    def __init__(self, patterns=replacement_patterns):
        self.patterns = [(re.compile(regex), repl) for (regex, repl) in patterns]
    
    def replace(self, text):
        s = text
        
        for (pattern, repl) in self.patterns:
            s = re.sub(pattern, repl, s)
        
        return s
```

```python
from replacers import RegexpReplacer
replacer = RegexpReplacer()

txt_corpus_replace = []
for song in songs['text']:
    song_replace = replacer.replace(song)
    txt_corpus_replace.append(song_replace)    
```

## Regular Expression

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
```

```python
txt_corpus_rmpunc = []
for song in txt_corpus_replace:
    song_rmpunc = tokenizer.tokenize(song)
    txt_corpus_rmpunc.append(song_rmpunc)
    

print(txt_corpus_rmpunc[0])
```

## Removing Stopwords (불용어 제거하기)
* nltk (natural language toolkit) 라이브러리 사용
* stopwords(불용어) -> 관사, 조사, 전치사, 접속사

```python
from nltk.corpus import stopwords
english_stops = set(stopwords.words('english'))

txt_corpus_nostopwords = []
for song in txt_corpus_rmpunc:
    word = [word for word in song if word not in english_stops]
    txt_corpus_nostopwords.append(word)
    
print(txt_corpus_nostopwords[0])
```

## Lemmatiziation (Lemma : 어원)
* print(stemmer.stem('believes')) -> believ
* print(lemmatizer.lemmatize('believes')) -> belief

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

txt_corpus_lemmas = []
for song in txt_corpus_nostopwords:
    txt_corpus_lemma = []    
    for song_word in song:
        word = lemmatizer.lemmatize(song_word, pos = 'n')
        txt_corpus_lemma.append(word)
        
    txt_corpus_lemmas.append(txt_corpus_lemma)
    
print(txt_corpus_lemmas[0])
```

```python
# Dimensionality of the resulting word vectors.
#more dimensions, more computationally expensive to train
#but also more accurate
#more dimensions = more generalized
num_features = 50
# Minimum word count threshold.
min_word_count = 1

# Number of threads to run in parallel.
#more workers, faster we train
num_workers = multiprocessing.cpu_count()

# Context window length.
context_size = 7

# Seed for the RNG, to make the results reproducible.
seed = 1

songs2vec = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
)

songs2vec.build_vocab(txt_corpus_lemmas)
print (len(txt_corpus_lemmas))
```

```python
import time
start_time = time.time()



songs2vec.train(txt_corpus_lemmas)

if not os.path.exists("trained"):
    os.makedirs("trained")

songs2vec.save(os.path.join("trained", "songs2vectors.w2v"))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import math
from textblob import TextBlob as tb

#tf 구하기
def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

#단어를 포함하는 document의 수 구하기
def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob.words)

#idf 구하기
def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

#tf-idf 구하기
def tfidf(word, blob, bloblist):
    return tf(word, blob) * idf(word, bloblist)
```

## Tf-Idf 생성 후 가수 별 top3 단어 추출

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.porter import PorterStemmer
```

```python
corpus = songs['text'].tolist()
tfidf = TfidfVectorizer(norm='l2', use_idf=True, smooth_idf=True, stop_words='english')
tfidf.fit_transform(corpus)

for artist, song in songs.groupby('artist'):
    feat = np.sum(tfidf.transform(song['text']).toarray(), axis=0)
    sort_idx = np.argsort(feat.flatten())[::-1] # Sort descending
    fav = [tfidf.get_feature_names()[idx] for idx in sort_idx.tolist()[:10]]
    print(artist + "'s Fav words:")
    for w in fav:
        print('\t' + w)
    print()
```
